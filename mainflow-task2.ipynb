```python
import pandas as pd
filePath = "HireData.csv"
dataframe  = pd.read_csv(filePath,encoding = 'latin1' )
```

```python
print("Original Data:")
dataframe.head()
```

```python
#1) Handling the missing values
#fill missing middle name with nothing 
dataframe.fillna({"MiddleName": "NULL"},inplace = True)
dataframe.head()
```

```python
dataframe.fillna({'ParentEmployeeKey':"noKey"}, inplace = True)
dataframe.head()
```

```python
dataframe['HireDate'] = pd.to_datetime(dataframe['HireDate'], errors='coerce')
dataframe['BirthDate'] = pd.to_datetime(dataframe['BirthDate'], errors='coerce')
dataframe.head()
```

```python
median_hire_date = dataframe['HireDate'].median()
dataframe.fillna( {'HireDate':median_hire_date}, inplace=True)
dataframe.head()
```

```python
current_date = pd.Timestamp.now()
dataframe['YearsOfService'] = (current_date - dataframe['HireDate']).dt.days // 365
```

```python
processed_file_path = 'AdvancedProcessedHireData.csv'
dataframe.to_csv(processed_file_path, index=False)
print(f"\nProcessed data saved to {processed_file_path}")
```
